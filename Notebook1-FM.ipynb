```python
#Calculamos la señal sinusoidal de la portadora
```

```python
muestras =frec_moduladora/mps
t=np.arange(0, 1, muestras)
```

```python
moduladora = np.sin(2 * np.pi * frec_moduladora * t)
#portadora *= 0.3  #atenuación
#portadora_ints = np.int16(portadora * 32767)
```

```python
señal_modulada_FM =np.sin(2 * np.pi * (frec_portadora+k*moduladora)*t)
```
